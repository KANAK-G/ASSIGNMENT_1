## Q3. Make a list of each Boolean operator's truth tables (i.e. every possible combination of Boolean values for the operator and what it evaluate ).

### Ans

```python
lst1 = [True,False,True,False]
print('list1','\t\t\t',lst1)

lst2 = [False,True,True,False]
print('list2','\t\t\t',lst2)

lst3 = []
for i in lst1:
    for j in lst2:
        j = j and i
        lst3.append(j)
print('list1 and list2','\t',lst3)

lst4 = []
for i in lst1:
    for j in lst2:
        j = i or j
        lst4.append(j)
print('list1 or list2','\t\t',lst4)
```

## Q10.How can you tell the difference between break and continue?
###  Ans. It will be more understandable by example.
### break - The break statement terminates the loop containing it. Control of the program flows to the statement immediately after the body of the loop.
###       If break statement is inside a nested loop (loop inside another loop), break will terminate the innermost loop.

```python
# Use of break statement inside loop

for s in "string":
    if s == "i":
        break
    print(s)

print("The end")
```

### continue : The continue statement is used to skip the rest of the code inside a loop for the current iteration only. 

### Loop does not terminate but continues on with the next iteration.



```python
# Program to show the use of continue statement inside loops

for val in "string":
    if val == "i":
        continue
    print(val)

print("The end")
```

## Q11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?
### Ans. There is no difference , all these three give the same output. All are the 3 ways to write one thing as we can see by our program.

```python
for i in range(10):
    print(i)
for j in range(0,10):
    print(j)
for k in range(0,10,1):
    print(k)
```
